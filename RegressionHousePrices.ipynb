```python
df = df.drop("id", axis=1)
```

```python
df['date'] = pd.to_datetime(df["date"])
```

```python
df["year"] = df["date"].apply(lambda date: date.year)
```

```python
df["month"] = df["date"].apply(lambda date: date.month)
```

```python
df = df.drop("date", axis=1)
```

```python
df = df.drop("zipcode", axis=1)
```

```python
X = df.drop("price", axis=1).values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
```

```python
model = Sequential()
model.add(Dense(19, activation="relu")) # 19 is the number of columns in the set
model.add(Dense(19, activation="relu"))
model.add(Dense(19, activation="relu"))
model.add(Dense(19, activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history) # compare train and test loss to evaluate overfitting
```

```python
losses.plot() # in the below plot, we can se that the loss is decreasing, as well as the validation loss
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df["price"].describe()
```

```python
explained_variance_score(y_test, predictions)
```
